Question 2:
The normalized distribution for a personâ€™s daily calorie intake has a mean of 30
calories and a standard deviation of 5 calories. Write a python code that Calculates and
visualizes the following probabilities for a randomly selected person having a daily
calorie intake:
I. Greater than or equal to 19 calories.
II. Less than or equal to 19 calories.
III. Between 19 and 35 calories.


```python
import matplotlib.pyplot as plt
import scipy.stats as stats

#  normal distribution data
samples = np.random.normal(60, 12, 60000)

# Plot histogram
plt.hist(samples, bins=30, density=True, color='red', alpha=0.6, edgecolor='black')

# Plot probability density function (PDF)
x = np.linspace(20, 100, 100)
plt.plot(x, stats.norm.pdf(x, 60, 12), 'r')

# Show the plot
plt.title("Histogram and PDF")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
# Question 2: Probability calculations
mean, std_dev = 30, 5
prob_greater_19 = 1 - stats.norm.cdf(19, mean, std_dev)
prob_less_19 = stats.norm.cdf(19, mean, std_dev)
prob_between_19_35 = stats.norm.cdf(35, mean, std_dev) - prob_less_19

print(f"P(X >= 19): {prob_greater_19:.4f}")
print(f"P(X <= 19): {prob_less_19:.4f}")
print(f"P(19 <= X <= 35): {prob_between_19_35:.4f}")
```
